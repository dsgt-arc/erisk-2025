```python
import pandas as pd
from collections import Counter
import nltk
from nltk.corpus import stopwords
from nltk.util import bigrams
import matplotlib.pyplot as plt
from wordcloud import WordCloud

# 🚀 Download stopwords if not already available
nltk.download('stopwords')

# Define a list of pronouns (from NLTK stopwords + manual additions)
pronouns = set([
    "i", "me", "my", "mine", "myself",
    "you", "your", "yours", "yourself", "yourselves",
    "he", "him", "his", "himself",
    "she", "her", "hers", "herself",
    "it", "its", "itself",
    "we", "us", "our", "ours", "ourselves",
    "they", "them", "their", "theirs", "themselves"
]) | set(stopwords.words('english'))  # Merge with standard stopwords

# 🚀 Stream word processing to avoid MemoryError
bigram_counts = Counter()

# Process text row-by-row for bigrams
for text in test["TEXT"].dropna():
    words = text.lower().split()  # Convert to lowercase and split words
    filtered_words = [word for word in words if word not in pronouns]  # Remove pronouns
    bigram_counts.update(bigrams(filtered_words))  # Extract bigrams and update counter

# Convert bigrams from tuple format to string format
bigram_freqs = {" ".join(bigram): count for bigram, count in bigram_counts.most_common(100)}  # Get top 100 bigrams

# Print the 20 most common bigrams
print("\n🔹 Most Common Bigrams (Excluding Pronouns) in TEXT Column:")
for bigram, count in list(bigram_freqs.items())[:20]:  # Print top 20 for reference
    print(f"{bigram}: {count:,}")

# 🚀 Generate Word Cloud for Bigrams
plt.figure(figsize=(10, 5))
wordcloud = WordCloud(width=800, height=400, background_color="white", 
                      colormap="viridis").generate_from_frequencies(bigram_freqs)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.title("Word Cloud of Most Common Bigrams (Excluding Pronouns)", fontsize=14)
plt.show()
```
